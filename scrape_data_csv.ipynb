```python
import pandas as pd
import os
import requests
import unittest
from unittest.mock import patch

def download_csv_levels(atom, output_dir):
    url = 'https://physics.nist.gov/cgi-bin/ASD/energy1.pl'
    params = {
        'de': 0,
        'spectrum': atom,
        'submit': 'Retrieve Data',
        'units': 0,
        'format': 2,  # choose format 2 for CSV output
        'output': 0,
        'page_size': 15,
        'multiplet_ordered': 0,
        'conf_out': 'on',
        'term_out': 'on',
        'level_out': 'on',
        'unc_out': 1,
        'j_out': 'on',
        'lande_out': 'on',
        'perc_out': 'on',
        'biblio': 'on',
        'temp': ''
    }

    response = requests.get(url, params=params)
    response.raise_for_status()  # raise an exception for bad responses

    # extract filename from response headers or generate a filename based on the atom
    filename = atom.replace(' ', '_') + '.csv'
    filepath = os.path.join(output_dir, filename)

    with open(filepath, 'wb') as f:
        f.write(response.content)

    print(f"Downloaded {filename}")



# read the Excel file containing the list of atoms
try:
    atoms_df_2 = pd.read_excel('singly_charged_atoms.xlsx')
    atoms = atoms_df_2['Atoms'].tolist()
except FileNotFoundError:
    print("Error: The Excel file was not found.")
    atoms = []
except Exception as e:
    print(f"Error: {e}")
    atoms = []

# read the Excel file containing the list of atoms
# atoms_df = pd.read_excel('neutral_atoms.xlsx')
atoms_df_2 = pd.read_excel('singly_charged_atoms.xlsx')

# extract the list of atoms from the DataFrame
atoms = atoms_df_2['Atoms'].tolist()

# directory to save the downloaded CSV files
output_directory = 'raw_singly_charged_data_levels'  

if not os.path.exists(output_directory):
    os.makedirs(output_directory)

for atom in atoms:
    download_csv_levels(atom, output_directory)


# Unit tests
class TestDownloadCSVLevels(unittest.TestCase):

    @patch('requests.get')
    def test_download_csv_levels(self, mock_get):
        # setup the mock response object
        mock_response = mock_get.return_value
        mock_response.status_code = 200
        mock_response.content = b'Test content'
        
        atom = 'H_I'
        output_dir = 'test_dir'
        os.makedirs(output_dir, exist_ok=True)
        
        # call the function
        download_csv_levels(atom, output_dir)
        
        # verify that the file was created and contains the expected content
        filename = os.path.join(output_dir, 'H_I.csv')
        with open(filename, 'rb') as f:
            self.assertEqual(f.read(), b'Test content')
        
        # clean up
        os.remove(filename)
        os.rmdir(output_dir)


```
